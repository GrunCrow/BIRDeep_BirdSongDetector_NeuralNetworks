```python
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
# Define the path to the audio file you want to process
audio_file = "../Dataset/Audios/AM4/2023_06_01/AM4_20230601_093000.WAV"
# Load the audio
y, sr = librosa.load(audio_file)

# Create the spectrogram without axis scales and color bar
plt.figure(figsize=(12, 6))  # Set the background color to black
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis=None, y_axis=None)

# Remove axis labels and color bar
plt.axis('off')

# Save the figure without a frame or border
plt.savefig('spectrogram.jpg', bbox_inches='tight', pad_inches=0, format='jpg')

plt.show()
```

```python
# Root directory where audio files are located
audio_root_directory = '../Dataset/Audios'  # Update to your audio directory
```

```python
# Function to create spectrogram images
def create_spectrogram(audio_file, output_folder):
    y, sr = librosa.load(audio_file)
    
    # Create the output path for the image
    output_image_path = audio_file.replace('Audios', 'Images')
    output_image_path = output_image_path.replace(".WAV", ".jpg")
    
    # Ensure the output folder exists
    os.makedirs(os.path.dirname(output_image_path), exist_ok=True)
    
    # Create the spectrogram without axis scales and color bar
    plt.figure(figsize=(12, 6))  # Set the background color to black
    D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis="time", y_axis="log")
    plt.axis('off')  # Remove axes
    plt.savefig(output_image_path, bbox_inches='tight', pad_inches=0)
    plt.close()

# Recursively iterate through all subdirectories and audio files
for root, dirs, files in os.walk(audio_root_directory):
    for file in files:
        if file.endswith('.WAV'):
            audio_file_path = os.path.join(root, file)
            create_spectrogram(audio_file_path, audio_root_directory)

# Define the progress bar
audio_files = [os.path.join(root, file) for root, dirs, files in os.walk(audio_root_directory) for file in files if file.endswith('.WAV')]
progress = tqdm(total=len(audio_files), unit="audio file", unit_scale=True)

# Recursively iterate through all subdirectories and audio files
for root, dirs, files in os.walk(audio_root_directory):
    for file in files:
        if file.endswith('.WAV'):
            audio_file_path = os.path.join(root, file)
            create_spectrogram(audio_file_path, audio_root_directory)
            progress.update(1)
progress.close()
```
